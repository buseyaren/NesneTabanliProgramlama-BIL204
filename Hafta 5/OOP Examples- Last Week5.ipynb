## Python Kapsülleme (Encapsulation)

Kapsülleme, nesne yönelimli programlamanın temel özelliklerinden biridir. Kapsülleme, niteliklerin ve yöntemlerin tek bir sınıf içinde paketlenmesini ifade eder.

Dış sınıfların bir sınıfın özniteliklerine ve yöntemlerine erişmesini ve bunları değiştirmesini engeller. Bu aynı zamanda veri gizlemenin sağlanmasına da yardımcı olur.

```python
class Computer:
   ## Data hiding: __constructor
    def __init__(self):
        self.__maxprice = 900  #maxprice artık gizlendi

    def sell(self):
        print("Selling Price: {}".format(self.__maxprice))

    def setMaxPrice(self, price): #price: 3000
        self.__maxprice = price #maxprice= 3000

c = Computer() #object created
c.sell()

# change the price
c.__maxprice = 3000 #price not updated, because __maxprice is hidden
#c.sell()


# using setter function - updated
c.setMaxPrice(3000)
c.sell()
```

Yukarıdaki programda bir Bilgisayar sınıfı tanımladık.

Bilgisayarın maksimum satış fiyatını saklamak için __init__() yöntemini kullandık. Burada, koda dikkat edin:

```python
c.__maxprice = 1000
```

Burada __maxprice değerini sınıf dışında değiştirmeye çalıştık. Ancak __maxprice özel bir değişken olduğu için çıktıda bu değişiklik görülmez.

Gösterildiği gibi, değeri değiştirmek için bir ayarlayıcı işlevi, yani parametre olarak fiyatı alan setMaxPrice() kullanmalıyız.

## Polimorfizm

Polimorfizm biyolojide, iki veya daha fazla farklı fenotipin aynı tür popülasyonunda bulunmasıdır. Terim kökeni bakımından Latince "çok" (poli) ve "biçim" (morfizmos) kelimelerinin bir araya gelmesiyle oluşmuştur. Başka bir deyişle, birden fazla biçimin bulunması olarak da tanımlanabilir. Bu şekilde sınıflandırılabilmek için, biçimlerin aynı zaman diliminde aynı habitatta bulunmaları gerekir.

Polymorphism, (çok biçimlilik) methodların objeye göre farklı çıktılar üretmesi veya farklı işler yapmasıdır. Yani alışageldiğimiz gibi methodlara sabit görevler vermek yerine onlara çok biçimli (polimorf) davranacak şekilde bir esneklik vermektir.


Polimorfizm, nesne yönelimli programlamanın bir başka önemli kavramıdır. Basitçe birden fazla form anlamına gelir.

Yani, aynı varlık (yöntem veya operatör veya nesne), farklı senaryolarda farklı işlemler gerçekleştirebilir.

Bir örnek görelim,

```python
#Inheritance
class Polygon:  #Çokgen- Şekil
    # method to render a shape
    def render(self):
        print("Rendering Polygon...")

class Square(Polygon):
    # renders Square
    def render(self):
        print("Square-Rendering Square...")

class Circle(Polygon):
    # renders circle
    def render(self):
        print("Polygon-Rendering Circle...")
    
# create an object of Square
p1 = Polygon()
p1.render()

s1 = Square()
s1.render()

s2 = Circle()
s2.render()
```

Yukarıdaki örnekte, bir üst sınıf oluşturduk: Polygon ve iki alt sınıf: Square ve Circle. 

render() yönteminin kullanımına dikkat edin.

render() yönteminin temel amacı şekli oluşturmaktır. Bununla birlikte, bir kare oluşturma işlemi, bir daire oluşturma işleminden farklıdır.

Bu nedenle, render() yöntemi farklı sınıflarda farklı davranır. Veya render()'ın polimorfik olduğunu söyleyebiliriz.

### Hatırlanması Gereken Önemli Noktalar:

* Nesne Yönelimli Programlama, programın verimli olduğu kadar anlaşılmasını da kolaylaştırır.
* Sınıf paylaşılabilir olduğu için kod yeniden kullanılabilir.
* Veri soyutlama ile veriler güvenli ve güvenlidir.
* Polimorfizm, farklı nesneler için aynı arayüze izin verir, böylece programcılar verimli kod yazabilir.

```python
class Animal:  #superclass

    # attribute and method of the parent class
    name = ""
    
    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):

    # new method in subclass
    def display(self):
        # access name attribute of superclass using self
        print("My name is", self.name)
        
class Monkey(Animal):
    
    def eatBanana(self):
        
        print("The monkey type ate the banana..")

# create an object of the subclass
labrador = Dog()
# access superclass attribute and method 
labrador.name = "Rohu"
#labrador.eat()

# call subclass method 
#labrador.display()

golden = Dog()
golden.name = "Ruby"
golden.display()
golden.eat()

chimpanzee = Monkey()
chimpanzee.eatBanana()
```

Burada, Animal sınıfının adına ve eat() öğesine erişmek için labrador (Köpeğin nesnesi) kullanıyoruz. Bu mümkündür çünkü alt sınıf, üst sınıfın tüm özniteliklerini ve yöntemlerini devralır.

Ayrıca, self kullanarak Dog sınıfının yöntemi içindeki name özniteliğine eriştik.

Polimorfizm, alt sınıfların ata sınıflardaki metotları geçersiz kılması(method overriding) sayesinde çok biçimli olarak davranmasına denir. Bu sayede alt sınıf ata sınıfından gelen davranışı kendine göre şekillendirebilir.

Metotlarda "Geçersiz Kılma" ise bir alt sınıfın içine doğrudan ya da dolaylı ata sınıflarından gelen bir(ya da daha fazla) yöntemin aynısının(aynı yöntem adı ve aynı parametre listesi) kodlanmasına verilen addır.

Polimorfizm sayesinde uygulamaların genişletilebilirliğini sağlarız ve bir ata sınıfın sunduğu yöntemleri geçersiz kılan alt sınıflar yardımı ile ata sınıfa göre kodlanmış tek bir kod kesimine farklı davranışlar yüklemek olanaklı olmaktadır. Öyleyse, elimizde esnek bir altyapı var demektir. Bu esneklik altyapıya yeni türlerin eklenmesi, kalıtım ve geçersiz kılma ilişkileri çerçevesinde oldukça kolaydır.

### Python'da kalıtım
Python'daki başka bir kalıtım örneğine bakalım,

Çokgen, 3 veya daha fazla kenarı olan kapalı bir şekildir. Diyelim ki aşağıdaki gibi tanımlanan Polygon adında bir sınıfımız var,

```python
class Polygon:
    def __init__(self, no_of_sides): 
        self.n = no_of_sides
        #print(self.n)
        self.sides = [0 for i in range(no_of_sides)] # list(array): for range - length
        #print(self.sides)
        #0-1-2-3-4-5 (Altıgen)
        
    def inputSides(self):
        self.sides = [float(input("Enter side "+str(i+1)+" : ")) 
                      for i in range(self.n)]
    def dispSides(self):
        for i in range(self.n): #n: input kenar sayısı
            print("Side",i+1,"is",self.sides[i])
```

```python
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self, 3)
t1 = Triangle()
t1.inputSides()
t1.dispSides()
```

Bu sınıf, kenar sayısı n'yi ve her bir kenarın büyüklüğünü taraflar adı verilen bir liste olarak depolamak için veri özniteliklerine sahiptir.

- inputSides() yöntemi, her bir tarafın büyüklüğünü alır
- dispSides() yöntemi, bu kenar uzunluklarını görüntüler
- Üçgen, 3 kenarı olan çokgendir. Böylece, Poligon'dan miras alan Üçgen adında bir sınıf oluşturabiliriz. Bu, Polygon sınıfının tüm özelliklerini Triangle sınıfı için kullanılabilir hale getirir.

Bunları yeniden tanımlamamıza gerek yok (kod yeniden kullanılabilirliği). Üçgen aşağıdaki gibi tanımlanabilir.

```python
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self,3)

    def findArea(self):
        a, b, c = self.sides
        # calculate the semi-perimeter
        s = (a + b + c) / 2
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('The area of the triangle is %0.2f' %area)
```

Ancak, Triangle sınıfı, üçgenin alanını bulmak ve yazdırmak için findArea() adlı yeni bir yönteme sahiptir.

Şimdi, bir nesne oluşturmak da dahil olmak üzere yukarıdaki örneğin tam çalışma kodunu görelim,

```python
class Polygon:
    # Initializing the number of sides
    def __init__(self, no_of_sides):
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def inputSides(self):
        self.sides = [float(input("Enter side "+str(i+1)+" : ")) for i in range(self.n)]

    # method to display the length of each side of the polygon
    def dispSides(self):
        for i in range(self.n):
            print("Side",i+1,"is",self.sides[i])

class Triangle(Polygon):
    # Initializing the number of sides of the triangle to 3 by 
    # calling the __init__ method of the Polygon class
    def __init__(self):
        Polygon.__init__(self,3)

    def findArea(self):
        a, b, c = self.sides

        # calculate the semi-perimeter
        s = (a + b + c) / 2

        # Using Heron's formula to calculate the area of the triangle
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('The area of the triangle is %0.2f' %area)

# Creating an instance of the Triangle class
t = Triangle()

# Prompting the user to enter the sides of the triangle
t.inputSides()

# Displaying the sides of the triangle
t.dispSides()

# Calculating and printing the area of the triangle
t.findArea()
```

Burada, Triangle sınıfı için ayrı ayrı inputSides() veya dispSides() gibi metotlar tanımlamasak da bunları kullanabildiğimizi görüyoruz.

Sınıfın kendisinde bir öznitelik bulunamazsa, arama temel sınıfa doğru devam eder. Temel sınıfın kendisi diğer sınıflardan türetilmişse, bu yinelemeli olarak tekrarlanır.

## Python Kalıtımında Yöntemi Geçersiz Kılma
Önceki örnekte, alt sınıfın nesnesinin üst sınıfın yöntemine erişebildiğini görüyoruz.

Ancak, aynı yöntem hem üst sınıfta hem de alt sınıfta mevcutsa ne olur?

Bu durumda, alt sınıftaki yöntem, üst sınıftaki yöntemi geçersiz kılar. Bu kavram, Python'da yöntem geçersiz kılma (override) olarak bilinir.

```python
class Animal: #superclass

    # attributes and method of the parent class
    name = "" #free

    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal): #subclass
    
    def __init__(self):
        pass
    # override eat() method
    def eat(self):
        print("I like to eat bones")

# create an object of the subclass
labrador = Dog()

# call the eat() method on the labrador object
labrador.eat()
```

Yukarıdaki örnekte, aynı eat() yöntemi hem Dog sınıfında hem de Animal sınıfında mevcuttur.

Şimdi, Dog alt sınıfının nesnesini kullanarak eat() yöntemini çağırdığımızda, Dog sınıfının yöntemi çağrılır.

Bunun nedeni, Dog alt sınıfının eat() yönteminin, Animal üst sınıfının aynı yöntemini geçersiz kılmasıdır.

## Python Kalıtımındaki super() Yöntemi
Daha önce, alt sınıftaki aynı yöntemin üst sınıftaki yöntemi geçersiz kıldığını gördük.

Ancak, alt sınıftan üst sınıf yöntemine erişmemiz gerekirse, super() yöntemini kullanırız. Örneğin,

```python
class Animal:

    name = ""
    
    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):
    
    # override eat() method
    def eat(self):   
        # call the eat() method of the superclass using super()
        super().eat()  # I can eat
        print("I like to eat bones") #I like to eat bones

# create an object of the subclass
labrador = Dog()

labrador.eat()
```

Yukarıdaki örnekte, Dog alt sınıfının eat() yöntemi, Animal üst sınıfının aynı yöntemini geçersiz kılar.

Dog sınıfının içinde, kullandık

```python
class Car():
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color
    
    def start_my_car(self):
        print('I am ready to drive!')
        
class Truck(Car):
    def __init__(self, brand, color, size):
        super().__init__(brand, color)
        self.size = size
        
    def start_my_car(self, key):
        if key == 'truck_key':
            print('I am ready to drive!')
        else:
            print('Key is not right')
            
    def stop_my_car(self, brake):
        if brake:
            print('The engine is stopped!')
        else:
            print('I am still running!')
            
truck1 = Truck('Toyota', 'Silver', 'Large')
truck1.start_my_car('truck_key')
truck1.stop_my_car(brake = False)
```

```python
#method overloading

import abc


class MyClass(object):

    __metaclass__ = abc.ABCMeta

    def my_set_val(self, value):
        self.value = value

    def my_get_val(self):
        return self.value

    @abc.abstractmethod
    def print_doc(self):
        return


class MyChildClass(MyClass):
    def my_set_val(self, value):
        if not isinstance(value, int):
            value = 0
        super(MyChildClass, self).my_set_val(self)

    def print_doc(self):
        print("Documentation for MyChild Class")


my_instance = MyChildClass()
my_instance.my_set_val(100)
print(my_instance.my_get_val())
print(my_instance.print_doc())
```

```python
#!/usr/bin/env python

# 34-abstractclasses-1.py

# This code snippet talks about Abstract Base Classes (abc).

# The `abc` module provides features to create
# Abstract Base Classes.

# To create an Abstract Base Class, set the `__metaclass__` magic method
# to `abc.ABCMeta`. This will mark the respective class as an Abstract
# Base Class.

# Now, in order to specify the methods which are to be enforced on the
# child classes, ie.. to create Abstract Methods, we use the decorator
# @abc.abstractmethod on the methods we need.

# The child class which inherits from an Abstract Base Class can implement
# methods of their own, but *should always* implement the methods defined in
# the parent ABC Class.

import abc


class My_ABC_Class(object):
    __metaclass__ = abc.ABCMeta

    @abc.abstractmethod
    def set_val(self, val):
        return

    @abc.abstractmethod
    def get_val(self):
        return


# Abstract Base Class defined above ^^^

# Custom class inheriting from the above Abstract Base Class, below


class MyClass(My_ABC_Class):
    def set_val(self, input):
        self.val = input

    def get_val(self):
        print("\nCalling the get_val() method")
        print("I'm part of the Abstract Methods defined in My_ABC_Class()")
        return self.val

    def hello(self):
        print("\nCalling the hello() method")
        print("I'm *not* part of the Abstract Methods defined in My_ABC_Class()")


my_class = MyClass()

my_class.set_val(10)
print(my_class.get_val())
my_class.hello()
```

```python
class SuperClass:

    def super_method(self):
        print("Super Class method called")

# define class that derive from SuperClass
class DerivedClass1(SuperClass):
    def derived1_method(self):
        print("Derived class 1 method called")

# define class that derive from DerivedClass1
class DerivedClass2(DerivedClass1):

    def derived2_method(self):
        print("Derived class 2 method called")

# create an object of DerivedClass2
d2 = DerivedClass2()

d2.super_method()  # Output: "Super Class method called"

d2.derived1_method()  # Output: "Derived class 1 method called"

d2.derived2_method()  # Output: "Derived class 2 method called"
```
